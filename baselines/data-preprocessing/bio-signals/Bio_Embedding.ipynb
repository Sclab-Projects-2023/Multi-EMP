```python
import pandas as pd
import numpy as np
from glob import glob
```

```python
def create_multifeature_sequences(annotation_files, data_dir):
    all_sequences = []
    for file in annotation_files:
        annotations = pd.read_csv(file)
        for vid_name in annotations['VID_NAME'].unique():
            participant_files = glob(f"{data_dir}/{vid_name}_Participant*.csv")
            sequences = []
            # Process data from each participant file
            for p_file in participant_files:
                print(participant_files)
                df = pd.read_csv(p_file)
                features = df[['eda_scl_usiemens', 'met', 'pulse_rate_bpm', 'temperature_celsius']].to_numpy()
                # Adjust the number of data points to 3
                if len(features) >= 3:
                    features = features[:3]  # Use only the first 3 data points
                else:
                    # If less than 3, pad the rest with zeros
                    padding = np.zeros((3 - len(features), features.shape[1]))
                    features = np.vstack((features, padding))
                
                sequences.append(features)
            # Calculate the average of sequences for each participant
            sequence = np.mean(sequences, axis=0)
            all_sequences.append(sequence)
    return np.array(all_sequences)

# Data preparation
data_dir = "Path"
train_sequences = create_multifeature_sequences(['CSV_PATH'], data_dir)

# Check the shape of train_sequences
print(train_sequences.shape)
```
